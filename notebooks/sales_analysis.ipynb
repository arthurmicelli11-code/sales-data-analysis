# Sales Data Analysis
**Project:** Sales Performance Analysis  
**Description:** Analysis of regional revenue, product performance, and sales trends using Python and SQL.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../data/sales_data.csv")
df.head()
```

## Data Overview
The dataset contains sales transactions including region, product, quantity, and unit price.


```python
df.info()
df.describe()
```

```python
df["revenue"] = df["quantity"] * df["unit_price"]
df.head()
```

## Revenue by Region
We analyze total revenue generated in each region.

```python
revenue_region = df.groupby("region")["revenue"].sum().sort_values(ascending=False)
revenue_region
```

```python
plt.figure()
revenue_region.plot(kind="bar")
plt.title("Total Revenue by Region")
plt.xlabel("Region")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("../visuals/revenue_by_region.png")
plt.show()
```

## Top Products by Revenue
Identify which products generate the most revenue.

```python
top_products = df.groupby("product")["revenue"].sum().sort_values(ascending=False)
top_products
```

```python
plt.figure()
top_products.plot(kind="bar")
plt.title("Top Products by Revenue")
plt.xlabel("Product")
plt.ylabel("Revenue")
plt.tight_layout()
plt.savefig("../visuals/top_products.png")
plt.show()
```

## Key Insights

- North America generates the highest total revenue.
- Electronics category drives most of the revenue.
- Laptops and Smartphones are top-performing products.
- Accessories contribute smaller but consistent revenue.
- Revenue distribution varies significantly by region.
